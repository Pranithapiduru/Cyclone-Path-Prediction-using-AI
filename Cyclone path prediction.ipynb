```python
# Step 2: Load the datasets (update paths if needed)
file_noaa = r"C:\Users\p.pranitha\OneDrive\Documents\cyclone path prediction using AI\DATASET\noaa_atlantic_hurricane.csv"
file_north_indian = r"C:\Users\p.pranitha\OneDrive\Documents\cyclone path prediction using AI\DATASET\seasonal_winter_cyclone_frequency.csv"
file_tropical = r"C:\Users\p.pranitha\OneDrive\Documents\cyclone path prediction using AI\DATASET\tropical_cyclone_tracks.csv"

df_noaa = pd.read_csv(file_noaa)
df_north_indian = pd.read_csv(file_north_indian)
df_tropical = pd.read_csv(file_tropical)

print("✅ Datasets loaded successfully!")
```

```python
# Step 5: Data Transformation

# ✅ Handle missing values (forward fill)
df_noaa.ffill(inplace=True)
df_north_indian.ffill(inplace=True)
df_tropical.ffill(inplace=True)

# ✅ Encode categorical columns if they exist
if 'Category' in df_noaa.columns:
    df_noaa['Category_Code'] = df_noaa['Category'].astype('category').cat.codes

if 'BASIN' in df_tropical.columns:
    df_tropical['BASIN_Code'] = df_tropical['BASIN'].astype('category').cat.codes

# ✅ Convert DATE column into Year/Month (if present)
if 'DATE' in df_noaa.columns:
    df_noaa['DATE'] = pd.to_datetime(df_noaa['DATE'], errors='coerce')
    df_noaa['Year'] = df_noaa['DATE'].dt.year
    df_noaa['Month'] = df_noaa['DATE'].dt.month

print("✅ Data transformation completed successfully!")
```

```python
# Select only numeric columns
numeric_noaa = df_noaa.select_dtypes(include=[np.number])

print("Numeric columns in NOAA dataset:\n", numeric_noaa.columns)

# Pick target column (change this if actual name differs)
target_col = None
for col in numeric_noaa.columns:
    if "wind" in col.lower():   # match column names like 'Wind', 'Wind_speed', etc.
        target_col = col
        break

if target_col:
    print(f"\n✅ Using '{target_col}' as target for correlation")
    
    # Correlation matrix
    plt.figure(figsize=(10,6))
    sns.heatmap(numeric_noaa.corr(), annot=True, cmap="coolwarm")
    plt.title("Correlation Heatmap - NOAA Dataset (Numeric Only)")
    plt.show()
    
    # Correlation with target
    correlations = numeric_noaa.corr()[target_col].sort_values(ascending=False)
    print("Correlation with target:\n", correlations)
    
    # Select strong features
    selected_features = correlations[correlations.abs() > 0.2].index.tolist()
    print("\n✅ Selected Features for Model Training:", selected_features)

else:
    print("⚠️ No wind-related column found in NOAA dataset.")
```
